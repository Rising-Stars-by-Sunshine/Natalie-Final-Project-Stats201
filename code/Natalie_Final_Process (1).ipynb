#Part 3

```python
df['POPg'] = df["Population"].pct_change(1)
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
cut = 0
```

```python
df['Positiveg'] = df['POPg'].apply(lambda x: 1 if x>= cut else 0)
df.head(40)
```

#Part 4

```python
df['POPg_past'] =df['POPg'].shift(1)
df.head()
```

```python
window = 5
```

```python
df['POPg_past_ma5']=df['POPg_past'].rolling(window=window,min_periods=1).mean()
df.head(20)
```

#Part 5

```python
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit()
for train_idx, test_idx in tss.split(df):
  pass

train_df = df.filter(items=train_idx,axis=0)
test_df = df.filter(items=test_idx,axis=0)
```
